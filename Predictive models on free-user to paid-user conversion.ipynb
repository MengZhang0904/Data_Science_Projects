# Introduction to the dataset
## users.csv
Teacher information from 2000 randomly selected schools in the database, and whether or not they created a quote

## viewings.csv 
Each of the teachers’ viewings of lessons during the 2016-2017 school year
* **previewed**: indicates whether a user has started two more videos 
* **potentially_taught**: true if at least 25 minutes of the lesson was watched during school hours and at least 30% of the video for that viewing. Internally, we count a potentially taught lesson as one hour of teaching. 
* **full_screened**: indicates whether a user used full screen mode when viewing the lesson

* Grade factor: Mark whether a teacher teaching grade 2 and lower, grades 3 to 5, or grade 6 and upper

```python
US_school_user.grade_level.unique()
```

```python
def grade2_lower(grade_level):
    grade_list = grade_level.split(',')
    lower_grades = ['Prep','Pre','K','1','2']
    check = 0
    if any(x in grade_list for x in lower_grades):
        check = 1
    return (check)

def grade3_5(grade_level):
    grade_list = grade_level.split(',')
    mid_grades = ['3','4','5']
    check = 0
    if any(x in grade_list for x in mid_grades):
        check = 1
    return (check)

def grade6_upper(grade_level):
    grade_list = grade_level.split(',')
    upper_grades = ['6','Upper']
    check = 0
    if any(x in grade_list for x in upper_grades):
        check = 1
    return (check)
```

```python
US_school_user_drop1['grade2_lower'] = US_school_user_drop1['grade_level'].apply(grade2_lower)
US_school_user_drop1['grade3_5'] = US_school_user_drop1['grade_level'].apply(grade3_5)
US_school_user_drop1['grade6_upper'] = US_school_user_drop1['grade_level'].apply(grade6_upper)
```

#### Approach to a predictive model with the variable state
* State is a high cardinality categorical variable, one-hot encoding may significantly increase the number of dimensions
* Use count to encode the category variable 'state'

```python
def count_encode(X, categorical_features, normalize=False):
    print('Count encoding: {}'.format(categorical_features))
    X_ = pd.DataFrame()
    for cat_feature in categorical_features:
        X_[cat_feature] = X[cat_feature].astype(
            'object').map(X[cat_feature].value_counts())
        if normalize:
            X_[cat_feature] = X_[cat_feature] / np.max(X_[cat_feature])
    X_ = X_.add_suffix('_count_encoded')
    if normalize:
        X_ = X_.astype(np.float32)
        X_ = X_.add_suffix('_normalized')
    else:
        X_ = X_.astype(np.uint32)
    return X_
```
