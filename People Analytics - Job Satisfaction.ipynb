# Introduction
## The Problem
The CEO of Vandelay Industries is interested in insights pull out of their data on employee survey and demographic info. Particularly, the CEO expects the following questions to be addressed:
* What survey items/constructs have the greatest impact on job satisfaction?
* What areas should we focus on for improvement? Is there anything else the CEO should know about those areas?
* Given the areas identified for improvement, what are some recommended actions or initiatives? Beside nudges for employees, what are some concrete recommendations for Leadership/HR on improvement areas?

## The Datasets
### The Survey Data <br>
It records employees' response to a survey covering the following constructs:
* Autonomy
* Self-Efficacy
* Supervisor support
* Task significance
* Job satisfaction
* Fair procedures
* Organizational authenticity
* Job embeddedness
* Coworker relationships <br>

Employees respond to each item on a 1 to 7 likert scale, with 1 indicating strongly disagree and 7 indicating strongly agree.

### The Demographic Information
It records the following information of employees:
* Gender
* Highest Education
* Region
* Manager or not
* Tenure
* Hourly paid or not
* Income
* Department <br>

Employees' responses are categorized.

As indicated by 99, i.e., N/A, the following items contain missing values: v6,v7,v9,v10,v15,v16,v17,v18,v20.<br>
Now I want to look at the proportions of missing values.

```python
prop_missing = []
for item in ['v6','v7','v9','v10','v15','v16','v17','v18','v20']:
    prop = len(survey_raw_df[survey_raw_df[item]==99])/len(survey_raw_df)
    print ('{} misses {}% values'.format(item, round(prop*100,4)))
    prop_missing.append(prop)
```

```python
ID_missing = []
for item in ['v6','v7','v9','v10','v15','v16','v17','v18','v20']:
    IDs = survey_raw_df[survey_raw_df[item]==99]['ID'].tolist()
    for i in IDs:
        if i not in ID_missing:
            ID_missing.append(i)
ID_missing
```

These are 14, about 3% of the employees, who at least has 1 missing value in their response to the survey. <br>
What percentage of items that each of the above employees miss to respond?

```python
survey_missing = survey_raw_df[survey_raw_df.ID.isin(ID_missing)]
pct_missing = []
for index,row in survey_missing.iterrows():
    cnt = sum(row == 99)
    pct = cnt/23.0
    print ('ID {} misses {} items, i.e., {}%'.format(row.ID, cnt, round(pct*100,4)))
```

```python
survey_missing.loc[:,['ID','v6','v7','v9','v10','v15','v16','v17','v18','v20']]
```

Items v9 and v6 have the highest missing rate and ID 70 missed the most items. <br>
I want to find out if the values are missing completely at random.

```python
# Import a package implementing the Little's MCAR test
from mcar_tests import *

# Preprocess the dataframe so that it fits the requirement of the mcar_test
processed_survey = survey_raw_df.replace(99,np.NaN)
processed_survey = processed_survey.set_index(['ID'])

# Run the test
mcar_res = McarTests(processed_survey)
```

```python
# Check the p-value of chi-squre test in the mcar_test
print('The p-value is ', mcar_res.mcar_test())
```

The p-value is very small, indicating the evidence is significant for me to reject the null hypothesis that the missingness mechanism of the incomplete dataset is Missing Completely At Random (MCAR). <br>
Further, I want to explore MCAR tests for each pair of variables.

```python
# Output a MCAR matrix. Values are True (missingness in index variable is MCAR for column variable) or False 
# (missingness in index variable is not MCAR for column variable)
mcar_mat = mcar_res.mcar_t_tests()

# Return a matrx showing variables that are not MCAR 
mcar_mat[mcar_mat.sum(axis=1)>1]
```

Now I narrowed down the variables that miss values non-completely random. Different remedies can be considered as I proceed.

## Scale and Aggregate Data
* Scale employees' responses to the survey into standard score
* Calculate construct score by taking the average of its item score 

```python
# Calculate the standard score for each item
survey_items = survey_raw_df.drop(columns='ID').columns.tolist()
for item in survey_items:
    col_mean = merged_data[item].mean()
    col_std = merged_data[item].std()
    merged_data[item] = merged_data[item].apply(lambda x: np.NaN if pd.isnull(x) else (x-col_mean)/col_std)

merged_data.iloc[0:5,-23:]
```

Now as you can see, the survey data are standardized.

```python
merged_data.to_csv('./Data/merged_data_standadized.csv')
```

```python
# Calculate the construct scores
merged_data['autonomy'] = merged_data.v1
merged_data['self_efficacy'] = merged_data[['v2','v3','v4']].mean(axis=1)
merged_data['supervisor_support'] = merged_data[['v5','v6','v7']].mean(axis=1)
merged_data['task_significance'] = merged_data[['v8','v9','v10']].mean(axis=1)
merged_data['job_satisfaction'] = merged_data[['v11','v12','v13','v14']].mean(axis=1)
merged_data['fair_procedures'] = merged_data[['v15','v16']].mean(axis=1)
merged_data['org_authenticity'] = merged_data[['v17','v18']].mean(axis=1)
merged_data['job_embeddedness'] = merged_data[['v19','v20']].mean(axis=1)
merged_data['coworker_relation'] = merged_data[['v21','v22','v23']].mean(axis=1)
```

```python
def corr_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(st.pearsonr(df[r], df[c])[1], 4)
    return pvalues
```

# Statistical Modeling
## Regress Job Satisfaction on Other Constructs
Both DV and IV are continuous variables, therefore OLS regression is used here.

```python
merged_data.rename(columns={'self-efficacy':'self_efficacy'},inplace=True)
merged_data.to_csv('./Data/merged_standard_construct.csv')
```

```python
construct_model = smf.ols('job_satisfaction~autonomy+self_efficacy+supervisor_support+task_significance+\
                          fair_procedures+org_authenticity+job_embeddedness+coworker_relation',data = merged_data).fit()
print(construct_model.summary())
```

The construct_model performs fine. All the other constructs have explained 83.1% variance of job satisfaction. Looking into the coefficient for each construct, they are all positive and the coefficients for autonomy and job_embeddedness are not significant. Consider dropping them in the following up analyses. 

## Regress Job Satisfaction on Demographic Variables

```python
merged_data.replace('Installation,Maintenance,Repair','Instal,Maintenance,Repair',inplace=True)
merged_data.replace('Transportation and Material Moving','Transportation,Material Moving',inplace=True)
merged_data.replace('Business and Financial Operations','Business,Financial Operations',inplace=True)
merged_data.to_csv('./Data/merged_standard_construct.csv')
```

```python
demographic_model = smf.ols('job_satisfaction~C(gender)+C(highest_ed)+C(region)+C(manager)+C(tenure)\
                             +C(hourly_salaried)+C(income)+C(department)',data = merged_data).fit()
print(demographic_model.summary())
```

The demographic_model only explained 13.5% variance of job satisfaction. Looking into the coefficient for each variable, significances exist for highest education, manager status, tenure and income. Consider investigate these in the following up analyses.

```python
demographic_aov_tab = anova_lm(demographic_model,typ=2)
print(demographic_aov_tab.round(4))
```

The above ANOVA table shows manager status and tenure have significant effects on job satisfaction, and highest education has a marginal significant effect on job satisfaction.

Because highest education, tenure, and income are ordinal variables. For exploration purpose, I want to check there are linear relationships between these three demographic variables and job satisfaction.

```python
demographic_model2 = smf.ols('job_satisfaction~C(gender)+highest_ed+C(region)+C(manager)+tenure\
                             +C(hourly_salaried)+income+C(department)',data = merged_data).fit()
print(demographic_model2.summary())
```

The above result still shows a positive effect of manager status on job satisfaction. Managers have a higher job satisfaction than non-managers. There is also a significant positive relationship between income and job satisfaction, a marginal negative relationship between highest education and job satisfaction.

The above explorations seem to suggest that manager status, income, tenure, and highest education relate to employees' job satisfaction. I want to check if there any interactions between these variables.

```python
demographic_model3 = smf.ols('job_satisfaction~highest_ed*C(manager)+income*C(manager)+tenure*C(manager)',
                             data = merged_data).fit()
print(demographic_model3.summary())
```

```python
demographic_aov_tab3 = anova_lm(demographic_model3,typ=2)
print(demographic_aov_tab3.round(4))
```

There are no interactions between manager status and highest education, income, tenure. Maybe because being a manager is positively associated with higher education, income and longer tenure.

## Regress Job Satisfaction on Survey Constructs and Demographic Variables
Based on results from the two models above, for the sake of time, the combined model considers all survey constructs except autonomy and job embeddedness (because their coefficients are not significant in the construct_model) and demographic variables like highest education, manager status, tenure and income.

```python
combined_model = smf.ols('job_satisfaction~self_efficacy+supervisor_support+task_significance+fair_procedures\
                          +org_authenticity+coworker_relation+C(highest_ed)+C(manager)+C(tenure)+C(income)',
                         data = merged_data).fit()
print(combined_model.summary())
```

```python
combined_aov_tab = anova_lm(combined_model,typ=2)
print(combined_aov_tab.round(4))
```

Both regression result and ANOVA table show when incorporating both survey constructs and demographic variables, the four demongraphic variables become insignificant.

## MANOVA of Demographic Variables and Survey Constructs

```python
maov = MANOVA.from_formula('job_satisfaction+autonomy+self_efficacy+supervisor_support+task_significance+\
                          fair_procedures+org_authenticity+job_embeddedness+coworker_relation  ~ highest_ed\
                          +manager+tenure+income', data=merged_data)
print(maov.mv_test())
```

# Recursive Feature Elimination
Consider survey and demographic variables in linear regression and use recursive feature elimination to find the important features.
## Construct Level

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
rfe_construct_df = pd.concat([merged_data.iloc[:,:9],merged_data.iloc[:,-9:]], axis=1).dropna()
```

```python
X_rfe_construct = rfe_construct_df.drop(columns=['ID','job_satisfaction'])
X_rfe_construct_dum = pd.get_dummies(X_rfe_construct,columns=['region','manager','department'])
y_rfe_construct = rfe_construct_df.job_satisfaction
```

```python
lr = LinearRegression()
rfe = RFE(lr, n_features_to_select=1)
rfe.fit(X_rfe_construct_dum,y_rfe_construct)

print ("Features sorted by their rank:")
print (sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), X_rfe_construct_dum.columns)))
```

Drop department

```python
X_rfe_construct2 = rfe_construct_df.drop(columns=['ID','job_satisfaction','department'])
X_rfe_construct_dum2 = pd.get_dummies(X_rfe_construct2,columns=['region','manager'])
```

```python
rfe.fit(X_rfe_construct_dum2,y_rfe_construct)

print ("Features sorted by their rank:")
print (sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), X_rfe_construct_dum2.columns)))
```
